```python
import scrapy
from scrapy.crawler import CrawlerProcess

class SpecSpider(scrapy.Spider):
    name = 'Spec'
    
    start_urls = [
        'https://www.spectator.co.uk/?nogeoredir=1',
    ]
    
    def parse(self, response):
        
        for div in response.xpath('//div[contains(@class, "card__text")]'):
            
            title1 = list(div.xpath('h2/a[@class="card__title-link"]/text()').extract())
            title2 = list(div.xpath('h3/a[@class="card__title-link"]/text()').extract())
            
            title = title1
            if not title:
                title = title2
            
            author1 = list(div.xpath('div/div[contains(@class, "meta__author")]/a/text()').extract())
            author2 = list(div.xpath('div[contains(@class, "meta__author")]/a/text()').extract())
            author3 = list(div.xpath('div/div[contains(@class, "meta__author")]/div/a/text()').extract())
            
            author = author1
            if not author:
                author = author2
            if not author:
                author = author3
                
            yield {
                'title': title,
                'author': author
            }
            
            
process = CrawlerProcess({
    'FEED_FORMAT': 'json',
    'FEED_URI': 'specfrontpage.json',
    'LOG_ENABLED': False
})

process.crawl(SpecSpider)
process.start()
print('Done')
```
