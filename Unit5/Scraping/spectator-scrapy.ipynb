```python
import scrapy
from scrapy.crawler import CrawlerProcess

class SpecSpider(scrapy.Spider):
    name = 'Spec'
    
    start_urls = [
        'https://www.spectator.co.uk/?nogeoredir=1',
    ]
    
    def parse(self, response):
        
        for div in response.xpath('//div'):
            
            if div.xpath('div[contains(@class, "card")]/div/h2/a/text()').extract() == None:
                continue
            else:            
                yield {
                    'title': div.xpath('div[contains(@class, "card")]/div/h2/a/text()').extract()
                }
            
process = CrawlerProcess({
    'FEED_FORMAT': 'json',
    'FEED_URI': 'specfrontpage.json',
    'LOG_ENABLED': False
})

process.crawl(SpecSpider)
process.start()
print('Done')
```

```python
import scrapy
from scrapy.crawler import CrawlerProcess

class SpecSpider(scrapy.Spider):
    name = 'Spec'
    
    start_urls = [
        'https://www.spectator.co.uk/?nogeoredir=1',
    ]
    
    def parse(self, response):
        
        for div in response.xpath('//div'):
            
            output = list(div.xpath('div[contains(@class, "card")]/div/h2/a/text()').extract())
            if output: 
                for i in output:
                    yield {
                        'title': i
                    }
            
            
            
process = CrawlerProcess({
    'FEED_FORMAT': 'json',
    'FEED_URI': 'specfrontpage.json',
    'LOG_ENABLED': False
})

process.crawl(SpecSpider)
process.start()
print('Done')
```
